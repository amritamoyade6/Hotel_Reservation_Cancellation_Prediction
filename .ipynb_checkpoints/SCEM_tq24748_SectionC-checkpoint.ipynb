**Encoding variables**

**Splitting the dataset**

```python
# Function to split data into train, test and validate

def split_data(df, target_column, test_size=0.2, val_size=0.25, random_state=42):

    train_val_df, test_df = train_test_split(df, test_size=test_size, random_state=random_state, stratify=df[target_column])
    
    train_df, val_df = train_test_split(train_val_df, test_size=val_size, random_state=random_state, stratify=train_val_df[target_column])
    
    return train_df, val_df, test_df
```

**Splitting dataset into 60:20:20 split**

```python
train_df, val_df, test_df = split_data(hotel_df_model, target_column="booking_status", test_size=0.2, val_size=0.25)

print(f"Training set size: {len(train_df)}")
print(f"Validation set size: {len(val_df)}")
print(f"Test set size: {len(test_df)}")
```

```python
X = hotel_df_model.drop(columns=['booking_status'])
y = hotel_df_model['booking_status']

X_train, X_val, y_train, y_val = train_df.drop(columns=['booking_status']), val_df.drop(columns=['booking_status']), train_df['booking_status'], val_df['booking_status']
```

```python
# Function to train and evaluate models on training data

def train_and_evaluate_model(model, X_train, y_train, X_val, y_val):
    
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_val_pred = model.predict(X_val)
    
    y_train_prob = model.predict_proba(X_train)[:, 1]
    y_val_prob = model.predict_proba(X_val)[:, 1]
    
    metrics = {
        "Accuracy": (accuracy_score(y_train, y_train_pred), accuracy_score(y_val, y_val_pred)),
        "Precision": (precision_score(y_train, y_train_pred, average='weighted'), precision_score(y_val, y_val_pred, average='weighted')),
        "Recall": (recall_score(y_train, y_train_pred, average='weighted'), recall_score(y_val, y_val_pred, average='weighted')),
        "F1-Score": (f1_score(y_train, y_train_pred, average='weighted'), f1_score(y_val, y_val_pred, average='weighted')),
        "AUC": (roc_auc_score(y_train, y_train_prob), roc_auc_score(y_val, y_val_prob)),
    }
    
    return model, metrics
```

```python
def plot_training_metrics(metrics, title="Training Metrics Evaluation"):
    
    metric_names = list(metrics.keys())
    train_values = [values[0] for values in metrics.values()]
    val_values = [values[1] for values in metrics.values()]
    
    plt.figure(figsize=(10, 6))
    plt.plot(metric_names, train_values, marker='o', label='Training', color='#4a90e2', linewidth=2)
    plt.plot(metric_names, val_values, marker='o', label='Validation', color='#d9534f', linewidth=2)
    plt.title(title)
    plt.xlabel('Metrics')
    plt.ylabel('Score')
    plt.ylim(0.8, 1.05)
    plt.legend()
    plt.grid(visible=True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
```

```python
# Plotting Training Metrics

plot_training_metrics(metrics, title="Training Metrics Evaluation")
```

```python
# Bias-Variance Analysis Plot

metrics = ['recall', 'precision', 'f1']
colors = ['#4a90e2', '#d9534f', '#5cb85c']

plt.figure(figsize=(12, 8))

for metric, color in zip(metrics, colors):
    train_sizes, train_scores, val_scores = learning_curve(
        RandomForestClassifier(class_weight="balanced", random_state=42),
        X_train,
        y_train,
        cv=5,
        scoring=metric,
        train_sizes=np.linspace(0.1, 1.0, 10),
        random_state=42
    )
    
    train_scores_mean = np.mean(train_scores, axis=1)
    val_scores_mean = np.mean(val_scores, axis=1)
    
    plt.plot(train_sizes, train_scores_mean, 'o-', label=f"Training {metric.capitalize()}", color=color, alpha=0.7)
    plt.plot(train_sizes, val_scores_mean, 'o--', label=f"Validation {metric.capitalize()}", color=color)

plt.title("Bias-Variance Analysis")
plt.xlabel("Training Set Size")
plt.ylabel("Score")
plt.legend(loc="best")
plt.grid(visible=True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
X_test = test_df.drop(columns=['booking_status'])
y_test = test_df['booking_status']
```

```python
# Function to test and evaluate model

def test_and_evaluate_model(model, X_test, y_test, label_encoder):
    
    y_test_pred = model.predict(X_test)
    y_test_prob = model.predict_proba(X_test)[:, 1]
    
    accuracy = accuracy_score(y_test, y_test_pred)
    precision = precision_score(y_test, y_test_pred, average='weighted')
    recall = recall_score(y_test, y_test_pred, average='weighted')
    f1 = f1_score(y_test, y_test_pred, average='weighted')
    auc = roc_auc_score(y_test, y_test_prob)
    
    test_metrics = {"Metric": ["Accuracy", "Precision", "Recall", "F1-Score", "AUC"],
                    "Score": [accuracy, precision, recall, f1, auc]}
    test_metrics_df = pd.DataFrame(test_metrics)
    print("Test Data Metrics:")
    print(test_metrics_df.to_markdown(index=False))

    return y_test_pred, y_test_prob
```

```python
# Evaluating Metrics on Test Data

y_test_pred, y_test_prob = test_and_evaluate_model(first_model, X_test, y_test, label_encoder)
```

```python
# Function to Plot Confusion Matrix

def plot_confusion_matrix(y_true, y_pred, labels, title="Confusion Matrix"):
    
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

```python
# Plotting Confusion Matrix

y_test_pred_decoded = label_encoder.inverse_transform(y_test_pred)
y_test_decoded = label_encoder.inverse_transform(y_test)

plot_confusion_matrix(y_test_decoded, y_test_pred_decoded, labels=['Not_Canceled', 'Canceled'])
```

```python
# Function to plot ROC Curve

def plot_roc_curve(y_true, y_prob, title="ROC Curve"):
    
    fpr, tpr, thresholds = roc_curve(y_true, y_prob)
    auc_score = roc_auc_score(y_true, y_prob)
    
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, label=f'AUC = {auc_score:.4f}', linewidth=2, color='b')
    plt.plot([0, 1], [0, 1], 'k--', linewidth=1)  # Diagonal line for random guessing
    plt.title(title)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc='lower right')
    plt.grid(visible=True, linestyle='--', alpha=0.6)
    plt.show()
```

```python
# Plotting the ROC Curve

plot_roc_curve(y_test, y_test_prob, title="ROC Curve - Test Set")
```

```python
# Function to Plot Precision-Recall Curve

def plot_precision_recall_curve(y_true, y_prob, title="Precision-Recall Curve"):
    
    precision, recall, thresholds = precision_recall_curve(y_true, y_prob)
    
    plt.figure(figsize=(8, 6))
    plt.plot(recall, precision, marker='.', color='b', linewidth=2, label='Model')
    plt.title(title)
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.legend(loc='lower left')
    plt.grid(visible=True, linestyle='--', alpha=0.6)
    plt.show()
```

```python
# Plotting Precision-Recall Curve

plot_precision_recall_curve(y_test, y_test_prob, title="Precision-Recall Curve - Test Set")
```

**Performing Grid Search with Cross Validation**

```python
param_grid = {
    'n_estimators': [50, 75, 100, 125],
    'max_depth': [5, 10, 15],
    'min_samples_split': [2, 5, 7],
    'min_samples_leaf': [1, 2, 4]
}

# Performing Grid Search to Find Best Parameters
grid_search = GridSearchCV(
    RandomForestClassifier(class_weight='balanced', random_state=42),
    param_grid,
    cv=5,
    scoring='f1'
)

grid_search.fit(X_train, y_train)
best_params = grid_search.best_params_
print("Best Parameters:", best_params)

results = pd.DataFrame(grid_search.cv_results_)
```

```python
columns_to_display = [
    "param_n_estimators",
    "param_max_depth",
    "param_min_samples_split",
    "param_min_samples_leaf",
    "mean_test_score",
    "rank_test_score",
]
results_table = results[columns_to_display]

results_table.columns = [
    "n_estimators",
    "max_depth",
    "min_samples_split",
    "min_samples_leaf",
    "Mean F1-Score",
    "Rank",
]

results_table = results_table.sort_values(by="Rank")
top_results = results_table.head(5)

print("\nGrid Search Results (Tabular Format):")
print(top_results.to_markdown(index=False))
```
