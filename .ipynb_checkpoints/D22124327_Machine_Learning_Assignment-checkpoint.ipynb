## Experiment details

We are conducting an experiment to verify the claim that K-fold cross validation is a better data splitting approach than hold-out testing when there is less data available. We are using wine dataset from UCI repository https://archive.ics.uci.edu/ml/machine-learning-databases/wine/ to verify our hypothesis.

* We are importing the dataset and shuffling it to create randomness and remove bias.
* Then we are dividing it into subsets, starting with taking 20 records for first subset, and increasing the size of subset by taking 20 more records each time in a way subset 1 has 20 records, subset 2 has 40 records, subset 3 has 60 and so on. The subsets are created to test whether increasing amount of data subsequently in multiple iterations does affect the performance.
* Then we are training and testing different models on all subsets iteratively and comparing the accuracies for all subsets.
* We have performed scaling and stratification since the models such as SVM are sensitive to scale of input features and to make sure  that all classes are proportionally represented.
* Then we are training and testing all the models 100 times to get their holdout and cross validation accuracies and comparing it. For holdout, we are making a 70:30 train and test split and for cross validation,  5 folds are done.

We are comparing accuracy as our main metric as it can provide a general indication of how well a model is performing in terms of overall classification correctness. Accuracy is the ratio of correctly predicted instances to the total number of instances in the dataset. It is effective when the number of instances belonging to each class are roughly same i.e., dataset is balanced. However, since our dataset is a bit imbalanced, hence we are using stratification as mentioned above so that all classes are proportionally represented and this little bit of imbalance can be mitigated.

However, we have also calculated the precsion, recall and F1 scores for better understanding of performance of a model.

### Hypothesis

**H0**: If the data available is less, then K-fold cross validation approach **is NOT better** data splitting approach than hold-out testing.

**HA**: If the data available is less, then K-fold cross validation approach **is better** data splitting approach than hold-out testing.

#### Set the random seed for reproducibility

```python
np.random.seed(42)
```

#### Shuffle the dataset

We are shuffling the dataset randomly to introduce randomness and ensure that the data is randomly distributed across subsets. It helps avoid overfitting issue and improves generalisation of a model.

```python
wine_df = wine_df.sample(frac=1).reset_index(drop=True)
```

#### Initializing the subset size to 20

We are initializing the subset with 20 records so that the dataset can be equally distributed in 8 subsets as it has 178 records

```python
subset_size = 20
```

#### Initialize the number of subsets to create i.e., 8 subsets

```python
num_subsets = int(wine_df.shape[0] / subset_size)
print("Number of subsets created: ", num_subsets)
```

#### Initialize minmax scaler

We are scaling the input data since the models such as SVM are sensitive to scale of input features. MinMax scaler makes it easier to make comparison as well, unlike of StandardScaler.

```python
scaler = MinMaxScaler()
```

### Create the subsets and train and test GaussianNB model using holdout and k-fold cross validation

```python
holdout_accuracy_scores = []
cv_accuracy_scores = []
holdout_precision_scores = []
cv_precision_scores = []
holdout_recall_scores = []
cv_recall_scores = []
holdout_f1_scores = []
cv_f1_scores = []

nb = GaussianNB()

# Create subplots for confusion matrices
fig, axes = plt.subplots(nrows=num_subsets, ncols=2, figsize=(12, num_subsets*6))
plt.subplots_adjust(hspace=0.4)

for i in range(num_subsets):
    print(f'Subset {i+1}/{num_subsets}')
    
    subset = wine_df.iloc[i*subset_size:(i+1)*subset_size]


    #Splitting X and y
    X = subset.drop('Type_of_wine', axis=1)
    y = subset['Type_of_wine']

    # Split into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y) # using stratification since 
                                                                                         # there is a class imbalance issue
                                                                                         # and stratification ensures 
                                                                                         # all classes are represented 
                                                                                         # proportionally


    # Normalize data
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Train Naive Bayes model
    nb.fit(X_train_scaled, y_train)

    # Test accuracy for holdout   
    nb_score = nb.score(X_test_scaled, y_test)
    print(f"Naive Bayes accuracy (holdout): {nb_score}")

    holdout_accuracy_scores.append(nb_score)  #Storing the holdout accuracy results in the holdout_accuracy_scores list
    
    # Calculate confusion matrix for holdout
    y_pred_holdout = nb.predict(X_test_scaled)
    cm_holdout = confusion_matrix(y_test, y_pred_holdout)
    
    # Plot confusion matrix for holdout
    sns.heatmap(cm_holdout, annot=True, fmt='d', cmap='Blues', ax=axes[i][0])
    axes[i][0].set_title(f'Confusion Matrix (Holdout) - Subset {i+1}')
    
   # Calculate precision, recall, and F1-score for holdout
    f1_score_holdout = f1_score(y_test, y_pred_holdout, average='weighted')
    precision_holdout = precision_score(y_test, y_pred_holdout, average='weighted')
    recall_holdout = recall_score(y_test, y_pred_holdout, average='weighted')
    
    holdout_precision_scores.append(precision_holdout)
    holdout_recall_scores.append(recall_holdout)
    holdout_f1_scores.append(f1_score_holdout)

    # Print precision, recall, and F1-score for holdout
    print(f"Precision (holdout): {precision_holdout} \n")
    print(f"Recall (holdout): {recall_holdout} \n")
    print(f"F1-score (holdout): {f1_score_holdout} \n")

    # Test accuracy for cross validation
    nb_scores = cross_val_score(nb, X_train_scaled, y_train, cv=5)
    nb_cv_score = nb_scores.mean()
    print(f"Naive Bayes accuracy (cross-validation): {nb_cv_score}")

    cv_accuracy_scores.append(nb_cv_score)   #Storing the cross validation accuracy results in the cv_accuracy_scores list

    # Calculate confusion matrix for cross-validation
    y_pred_cv = cross_val_predict(nb, X_train_scaled, y_train, cv=5)
    cm_cv = confusion_matrix(y_train, y_pred_cv)

    # Plot confusion matrix for cross-validation
    sns.heatmap(cm_cv, annot=True, fmt='d', cmap='Blues', ax=axes[i][1])
    axes[i][1].set_title(f'Confusion Matrix (Cross-validation) - Subset {i+1}')

    # Calculate precision, recall, and F1-score for cross-validation
    precision_cv, recall_cv, f1_score_cv, _ = precision_recall_fscore_support(y_train, y_pred_cv, average='weighted')
    
    cv_precision_scores.append(precision_cv)
    cv_recall_scores.append(recall_cv)
    cv_f1_scores.append(f1_score_cv)
    
    # Print precision, recall, and F1-score for holdout
    print(f"Precision (CV): {precision_cv} \n")
    print(f"Recall (CV): {recall_cv} \n")
    print(f"F1-score (CV): {f1_score_cv} \n")

plt.show()
```

### Comparing the accuracy Holdout vs Cross Validation - Naive Bayes

Plotting a line graph to compare the accuracy of holdout vs cross validation across 8 subsets.

```python
# Plot holdout and cross-validation accuracy and F1-score separately
plt.figure(figsize=(10, 6))
plt.plot(range(1, num_subsets + 1), holdout_accuracy_scores, label='Holdout Accuracy')
plt.plot(range(1, num_subsets + 1), cv_accuracy_scores, label='Cross-validation Accuracy')
plt.xlabel('Subset')
plt.ylabel('Score')
plt.title('Holdout vs. Cross-validation Accuracy (Naive Bayes)')
plt.legend()
plt.show()
```

Cross validation clearly has a relatively consistent performance across all the subsets and has better performance when using subsets with less data.

### Comparing the accuracy performance of different models over 100 iterations

```python
# Define the models
models = [nb, rf, svm, knn]

# Number of iterations
n = 100

# Create arrays to store results
hold_out_acc = np.zeros(n)
cf_acc = np.zeros(n)

# Loop over models
for model in models:
    # Loop over iterations
    for i in range(n):
        # Split the dataset into training and testing sets for hold-out validation
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=i, stratify=y)

        # Calculate the accuracy of the hold-out validation
        model.fit(X_train, y_train)
        hold_out_acc[i] = model.score(X_test, y_test)

        # Calculate the accuracy of the k-fold cross-validation
        cf_acc[i] = cross_val_score(model, X, y, cv=5, scoring='accuracy').mean()

    # Create a line plot showing the performance metrics of the two validation methods over 10 iterations
    plt.plot(hold_out_acc, label='Hold-out validation')
    plt.plot(cf_acc, label='k-fold cross-validation')
    plt.xlabel('Iteration')
    plt.ylabel('Accuracy')
    plt.title('Hold-out validation vs. k-fold cross-validation over {} iterations for {}'.format(n, type(model).__name__))
    plt.legend()
    plt.show()
```

As per above graphs, it's evident that k-fold cross validation has a consistent performance over 100 iterations, however the hold out validation has a very jittery graph i.e., has very inconsistent performance.

## Conclusion

As evident from the experiment, k-fold cross validation technique has better stability and performance than the holdout method, which is also very prone to overfitting, when less data is present. Hence, we can accept the alternative hypothesis that states **"If the data available is less, then K-fold cross validation approach is better data splitting approach than hold-out testing."**
